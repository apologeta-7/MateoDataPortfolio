# Source Inference Refactor

This notebook is a handoff note for the final inference refactor. The executable serving path lives in `src/`, `api/` and `tests/` instead of notebook cells.

## Refactored Modules

- `src/config/paths.py`: project path discovery.
- `src/config/inference.py`: inference constants and artifact paths.
- `src/inference/artifacts.py`: artifact inventory, feature contract and serving model selection.
- `src/inference/model.py`: LightGBM loading and contract validation.
- `src/inference/validation.py`: input, feature and output validation.
- `src/inference/predict.py`: log-scale prediction, inverse transform and clipping.
- `src/inference/postprocess.py`: output and submission helpers.
- `api/`: FastAPI serving layer.
